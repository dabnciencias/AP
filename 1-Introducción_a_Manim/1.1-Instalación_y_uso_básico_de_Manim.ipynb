## Uso básico de Manim

Después de haber instalado `manim`, para poder usar este paquete en un _notebook_, siempre debemos empezar por cargar sus contenidos con la siguiente línea de código:

```python
from manim import * # Carga todos los contenidos del paquete 'manim'.
```

Ahora, podemos crear escenas de Manim ejecutando código como el del siguiente ejemplo:

```python
%%manim -qm -v WARNING MovingAround

class MovingAround(Scene):                         # Creamos una escena
    def construct(self):
        
        square = Square(color=BLUE, fill_opacity=1) # Creamos un cuadrado, lo
        self.play(square.animate.shift(LEFT))      # movemos a la izquierda,
        self.play(square.animate.set_fill(ORANGE))  # cambiamos su relleno,
        self.play(square.animate.scale(0.3))       # lo reescalamos y
        self.play(square.animate.rotate(0.4))      # lo rotamos.
```

Con esto, debe aparecer un recuadro negro con un cuadrado azul en medio. Dando clic dentro de este recuadro, podemos reproducir la animación de la escena. También podemos darle doble clic al recuadro para verla en pantalla completa.

En el ejemplo anterior, la escena que creamos se llama `MovingAround`, y fue creada con la línea de código

`class MovingAround(Scene):`.

Todo el código que está por debajo de la línea anterior describe qué sucederá en la escena. La línea de código que convierte todo este bloque de código en una animación es

`%%manim -qm -v WARNING MovingAround`,

donde
* ``%%manim`` nos permite ejecutar el comando `manim` desde un _notebook_ de Jupyter (desde una terminal virtual no es necesario poner los símbolos `%%` al inicio de la línea);
* `-qm` es una abreviación de _quality meduim_, lo que genera una animación de calidad media (si tarda mucho en generarse, podemos cambiarlo por `-ql`, _low quality_, que se genera más rápidamente);
* `-v WARNING` hace que `manim` sólo imprima mensajes de texto si son advertencias de que el código puede tener errores, en vez de imprimir el progreso del proceso de _rendering_;
* `MovingAround` indica el nombre de la escena que queremos animar, el cual siempre debe ir **después de haber llamado el comando** `manim` **y haber especificado todas las opciones que queremos para nuestra animación**.

### Opciones básicas de `manim` en la línea de comando

Los pedazos de código `-ql`, `-qm`, `-qh`, `-qk` y `-v` son ejemplos de opciones (o _flags_) que podemos darle al comando `manim` al momento de generar una escena. A continuación, veremos algunas de las opciones más relevantes; sin embargo, puedes consultar una lista completa de ellas en [esta liga](https://docs.manim.community/en/stable/tutorials/configuration.html?highlight=CLI%20options#a-list-of-all-cli-flags).

#### Opciones de calidad

Los primeros cuatro son ejemplos de _opciones de calidad_, pues determinan la calidad que tendrá la animación generada de la manera siguiente:

* `-ql` (_quality low_) calidad baja: resolución de 480p (640x480 pixeles);
* `-qm` (_quality medium_) calidad media: resolución de 720p (1080x720 pixeles);
* `-qh` (_quality high_) calidad alta: resolución de 1080p (1920x1080 pixeles);
* `-qk` (_quality 4k_) calidad 4k (3840 × 2160 pixeles).

#### Opciones de texto de mensajes

Por otro lado, `-v` es una abreviación de _verbose_; esta opción nos habla sobre qué tanta información queremos que nos devuelva el comando `manim` sobre el proceso que estará ejecutando. Como mencionamos anteriormente, la opción `-v WARNING` hace que `manim` sólo imprima mensajes de texto si son advertencias de que el código puede tener errores.

#### Opción de cuadros por segundo

El _flag_ `--fps` sirve para especificar el número de _frames per second_ (cuadros por segundo) de la animación. Por ejemplo, la opción

`--fps 60`

hará que `manim` genere una animación con 60 cuadros por segundo.

#### Opción de _output_

El _flag_ `-o` sirve para especificar el nombre del archivo que se generará con la animación. Por ejemplo, la opción

`-o mi_archivo.mp4`

hará que `manim` genere un archivo nombrado `mi_archivo.mp4` en la misma carpeta desde donde se ejecute el comando que contenga la animación correspondiente a la escena. También podemos determinar en dónde se ubicará el archivo generado escribiéndo el "nombre completo" (_path_) que queremos para él.

#### Opción _save last frame_

La opción `-s` hará que sólo se genere el último cuadro de animación de nuestra escena. Se puede usar en conjunción con alguna opción de calidad o en ausencia de ella, como en el siguiente ejemplo:

```python
%%manim -s -v WARNING MovingAround

# Usamos la opción -s para que sólo se genere el último cuadro de
# animación.

class MovingAround(Scene):                         # Creamos una escena
    def construct(self):
        
        square = Square(color=BLUE, fill_opacity=1) # Creamos un cuadrado, lo
        self.play(square.animate.shift(LEFT))      # movemos a la izquierda,
        self.play(square.animate.set_fill(ORANGE))  # cambiamos su relleno,
        self.play(square.animate.scale(0.3))       # lo reescalamos y
        self.play(square.animate.rotate(0.4))      # lo rotamos.
```

Esta opción es de gran utilidad cuando, por ejemplo, sólo queremos revisar la alineación de algunos objetos al final de la escena, para lo cual no vale la pena estar generando la animación completa varias veces y trasladándonos al final de ella.

#### Opción de prevista

El _flag_ `-p` sirve para que `manim` muestre el video con la animación una vez que haya terminado de generarla; sin embargo, **en Jupyter no es necesario utilizarla**.
